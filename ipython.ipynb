```python
import pandas as pd
import json
```

```python
df_categories = pd.read_csv('hs_categories_classification.csv', delimiter=';', dtype=str)
df_categories.head(5)
```

```python
categories_dict = df_categories.to_dict(orient='records')
```

```python
with open('hs_categories_classification.json', 'w') as file:
    json.dump(categories_dict, file, indent=4)
```

```python
df_products = pd.read_csv('hs_products_classification.csv', delimiter=';', dtype=str)
df_products.head(5)
```

```python
products_dict = df_products.to_dict(orient='records')
```

```python
with open('hs_products_classification.json', 'w') as file:
    json.dump(products_dict, file, indent=4)
```
